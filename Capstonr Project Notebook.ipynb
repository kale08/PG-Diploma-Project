# Normalizing values

```python
# Defining a fucntion to Normalize the price values

def normalised(l):
    n= []
    for i in l:
        n.append(i/l[0])
    return n
df_normalised = df_analysis.copy()
```

```python
for i in df_analysis.columns:
    df_normalised[i] = normalised(df_normalised[i])
```

# Daily Returns

```python
# Defining a fucntion to calculate daily returns of the stocks

def daily_returns_cal(l):
    n=[]
    for i in range(1,l.shape[0]):
        n.append(100*(l[i]/l[i-1] - 1))
    return(n)
```

```python
# Initializing a dataframe named 'Daily returns' 

daily_returns = pd.DataFrame(daily_returns_cal(df_analysis['S&P500']),columns=['S&P500'])
```

```python
for i in df_analysis.columns[1:]:
    daily_returns[i] = daily_returns_cal(df_analysis[i])
```

```python
daily_returns.head()
```

```python
daily_returns.to_csv('Daily Returns.csv') # Extracting data for visualisation
```

### Summary Statistics

```python
# Checking the statastics of the daily returns of stocks

summary_statistics = daily_returns.describe().loc[['mean','std','min','max']]
```

```python
summary_statistics.rename(index = {'mean':'Average Daily Returns','std':'Risk','min':'Minimum','max':'Maximum'},inplace= True)
```

```python
summary_statistics
```

#### Assumming the number of trading windows = 252

```python
# Calculating Annual Returns

Annual_Returns = pd.DataFrame(summary_statistics.loc['Average Daily Returns']*252)
Annual_Returns.rename(columns = {'Average Daily Returns':'Annual Return %'},inplace=True)
```

```python
# Calculating risk

Annual_Risk = pd.DataFrame(summary_statistics.loc['Risk']*(252**(1/2)))
```

```python
# Calculating Cummulative Returns

Cummulative_Returns = pd.DataFrame(100*(df_analysis.iloc[-1] - df_analysis.iloc[0])/df_analysis.iloc[0])
```

```python
# Combining the data of metrics

Descriptive_Metrics = pd.concat([Annual_Returns,Annual_Risk,Cummulative_Returns],axis=1)
```

```python
# Renaming the columns

Descriptive_Metrics.rename(columns={'Risk': 'Annual Risk %',0:'Cummulative Returns %'},inplace=True)
```

```python
Descriptive_Metrics
```

```python
Descriptive_Metrics['Sharpee Ratio'] = (Descriptive_Metrics['Annual Return %'] - 0.75)/Descriptive_Metrics['Annual Risk %']
```

```python
Descriptive_Metrics
```

```python
Descriptive_Metrics.to_csv('Metrics.csv') # Exporting the data for visualisation
```
